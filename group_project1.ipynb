```python
import pandas as pd
```

```python
def ask_name(s):
    """This function asks and checks if the answer contains only alphabetic characters"""
    while True:
        name = input(s)
        
        # Check if the entry is not empty
        if not name.strip():
            print("The name must not be empty.")
            continue
       
        # Check if the input contains only letters (no numbers or special characters)
        if not name.isalpha():
            print("The name must contain only letters.")
            continue
        # If all conditions are met, exit the loop
        break
    
    print(f"{name}, Good!")
    return name

def extract_tri_consonnes(s):
    """This function extracts three consonants and completes with 'X' when the entry does not have enough"""
    s = s.upper()  # Convert the string to uppercase
    vowels = "oeiyua".upper()  # List of vowels in uppercase
    consonants = ""
    a = 3
    char_list = list(s)  # Convert the string to a list of characters
    count = 0
    for char in char_list:
        if char in vowels or char == " ":
            pass
        else:
            consonants += char  # Add consonant to the result string
            count += 1
            if count == a:
                break
    if len(consonants) < a:
        for i in range(len(consonants), a):
            consonants = consonants + "X"  # Add 'X' to make up for missing consonants
    return consonants

def extract_tri_first_letter(s):
    """This function extracts the first three letters and completes with 'X' if less than three letters are available"""
    s = s.upper()  # Convert the string to uppercase
    a = 3
    list_of_letters = list(s)  # Convert the string to a list of characters
    tri_letters = ""
    if len(s) >= a:
        tri_letters = list_of_letters[0] + list_of_letters[1] + list_of_letters[2]
    else:
        tri_letters = s
        for i in range(len(s), a):
            tri_letters = tri_letters + "X"  # Add 'X' to make up for missing letters
    return tri_letters
    
def format_int(number):
    """This function formats the number with leading zeros to ensure it is three digits"""
    formatted_number = f'{number:03}'
    return formatted_number

def generateId(s1, s2, n=1):
    """This function generates an ID based on three consonants from the first string, 
    three first letters from the second string, and a formatted integer"""
    return  extract_tri_consonnes(s1) + extract_tri_first_letter(s2) +  format_int(n)

def add_row_to_excel(filename="registers.xlsx", new_row=None):
    """This function adds a new row to an Excel file. If the file does not exist, it creates a new one."""
    if new_row is None:
        return False
    try:
        # Read the existing Excel file
        df = pd.read_excel(filename)
        print(f"File {filename} loaded successfully.")
    except FileNotFoundError:
        # If the file does not exist, create a new DataFrame
        df = pd.DataFrame(columns=["ID", "Firstname", "Lastname"])
        print(f"File {filename} not found, creating a new file.")
    
    # Create a DataFrame for the new row
    new_row_df = pd.DataFrame([new_row])
    
    # Add the new row to the existing DataFrame with pd.concat()
    df = pd.concat([df, new_row_df], ignore_index=True)
    
    # Save back to the Excel file
    df.to_excel(filename, index=False)
    
    print(f"New row added and file {filename} saved.")
    return True

def incrementid(ID):
    """This function increments the numeric part of the ID"""
    n = int(ID[6:9]) + 1
    return f'{ID[0:6]}{n:03}'

def get_last_row_by_id(filename="Myfile.xlsx", id_to_check=None):
    """This function retrieves the last row with a specific ID from an Excel file"""
    if id_to_check is None:
        print("No ID provided for checking.")
        return None
    
    try:
        # Read the Excel file
        df = pd.read_excel(filename)
        print(f"File {filename} loaded successfully.")
        
        # Check if the 'ID' column exists in the file
        if 'ID' not in df.columns:
            print("The 'ID' column does not exist in this file.")
            return None
        
        # Filter the DataFrame to get the rows matching the ID
        filtered_df = df[df['ID'].str[1:5] == id_to_check[1:5]]
        
        if filtered_df.empty:
            print(f"The ID {id_to_check} does not exist in the file.")
            return None
        else:
            # Return the last row where the ID is found
            last_row = filtered_df.iloc[-1]
            return last_row
    
    except FileNotFoundError:
        print(f"The file {filename} was not found.")
        return None
```

```python
file_name = "registers.xlsx"

# Prompt the user to enter their first name and check the validity
firstname = ask_name("Enter your firstname ")

# Prompt the user to enter their last name and check the validity
lastname = ask_name("Enter your lastname ")

# Generate an ID based on the first three consonants of the first name, 
# the first three letters of the last name, and a default number 1
id = generateId(extract_tri_consonnes(lastname), extract_tri_first_letter(firstname), n = 1)

# Create a dictionary for the new row with ID, last name, and first name
new_row = {"ID": id, "Lastname": lastname, "Firstname": firstname}

# Check if the ID already exists in the file
if get_last_row_by_id(file_name, id) is None:
    # If the ID does not exist, add the new row to the Excel file
    add_row_to_excel(filename = "registers.xlsx", new_row = new_row)
    print("I am in the first condition")
else:
    # If the ID exists, get the last ID from the file
    last_id = get_last_row_by_id(file_name, id)["ID"]
    print(last_id)
    # Increment the last ID to generate a new unique ID
    new_id = incrementid(last_id)
    # Create a new row with the incremented ID
    new_row = {"ID": new_id, "Lastname": lastname, "Firstname": firstname}
    # Add the new row to the Excel file
    add_row_to_excel(filename = "registers.xlsx", new_row = new_row)

# Visualization of our file registers.xlsx
df = pd.read_excel(file_name)  # Read the Excel file
print(df)  # Print the DataFrame to visualize the content of the file
```
