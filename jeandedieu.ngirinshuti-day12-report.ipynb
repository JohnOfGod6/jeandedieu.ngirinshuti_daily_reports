### Exercise 9:
Create a subplot with 2 rows and 1 column. Plot `y1 = x^2` in the first subplot and `y2 = x^3` in the second.

````python
x = np.linspace(0,10)
y1 = x ** 2
y2 = x ** 3
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(6, 4))

# Plot y1 = x^2 in the first subplot
ax1.plot(x, y1, label='$y1 = x^2$')
ax1.set_title('Plot of $y1 = x^2$')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.grid()
ax1.legend()

# Plot y2 = x^3 in the second subplot
ax2.plot(x, y2, label='$y2 = x^3$' )
ax2.set_title('Plot of $y2 = x^3$')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.grid()
ax2.legend()

plt.tight_layout()

plt.show()

````

### Exercise 10:
Use a different color and linestyle for each line in a plot of three functions: `y1 = x`, `y2 = x^2`, and `y3 = x^3`.

````python
x = np.linspace(0,10)
y_1 = x
y_2 = x ** 2
y_3 = x ** 3
plt.plot(x, y_1, label='$y1 = x$', color = 'blue',linestyle = '-')
plt.plot(x, y_2, label='$y2 = x^2$', color = 'green', linestyle = '--')
plt.plot(x, y_3, label='$y3 = x^3$', color = 'red', linestyle = '-.')
plt.title('Line Plots of $y = x , y = x^2$ and $y = x^3$')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid()
plt.show()
````

````python
x = np.linspace(0,10)
y1 = x
y2 = x ** 2
y3 = x ** 3
fig, (ax1, ax2 , ax3) = plt.subplots(1, 3, figsize=(10, 4))

# Plot y1 = x in the first subplot
ax1.plot(x, y1, label='$y1 = x$', color = 'blue', linestyle = '-')
ax1.set_title('Plot of $y1 = x$')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.grid()
ax1.legend()

# Plot y2 = x^2 in the second subplot
ax2.plot(x, y2, label='$y2 = x^2$',  color = 'green',linestyle = '--')
ax2.set_title('Plot of $y2 = x^2$')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.grid()
ax2.legend()

# Plot y3 = x^3 in the third subplot
ax3.plot(x, y3, label='$y3 = x^3$', color = 'red', linestyle = '-.' )
ax3.set_title('Plot of $y3 = x^3$')
ax3.set_xlabel('x')
ax3.set_ylabel('y')
ax3.grid()
ax3.legend()

plt.tight_layout()

plt.show()

````

### Exercise 14:
Plot a filled area under the curve of `y = x^2` for `x` ranging from 0 to 10.
 

````python
 x = np.linspace(0, 10, 100)
y = x**2
plt.plot(x, y, label='$y = x^2$')
plt.fill_between(x, y, color='skyblue', alpha=0.4)
plt.title('Filled Area Under the Curve of $y = x^2$')
plt.xlabel('x')
plt.ylabel('$y = x^2$')
plt.legend()
plt.show()
````

### Exercise 16:
Create a stacked bar chart for the following data: `Group A = [3, 5, 7]`, `Group B = [2, 6, 4]`, and `categories = ['X', 'Y', 'Z']`.
 

````python
categories = ['X', 'Y', 'Z']
group_A = [3, 5, 7]
group_B = [2, 6, 4]
fig, ax = plt.subplots()
 
index = np.arange(len(categories))
bar1 = ax.bar(index, group_A,  label='Group A')
bar2 = ax.bar(index, group_B,  bottom=group_A, label='Group B')
ax.set_xlabel('Categories')
ax.set_ylabel('Values')
ax.set_title('Stacked Bar Chart')
ax.set_xticks(index)
ax.set_xticklabels(categories)
ax.legend()
plt.show()
````

### Exercise 17:
Create a plot with multiple subplots in a 2x2 grid. Plot `y1 = x`, `y2 = x^2`, `y3 = x^3`, and `y4 = x^4`.
 

````python
x = np.linspace(0, 10, 100)
y1 = x
y2 = x**2
y3 = x**3
y4 = x**4
 
fig, axs = plt.subplots(2, 2, figsize=(10, 8))
axs[0, 0].plot(x, y1, 'r')
axs[0, 0].set_title('y = x')

axs[0, 1].plot(x, y2, 'g')
axs[0, 1].set_title('y = x^2')

axs[1, 0].plot(x, y3, 'b')
axs[1, 0].set_title('y = x^3')

axs[1, 1].plot(x, y4, 'm')
axs[1, 1].set_title('y = x^4')
 
fig.suptitle('Multiple Subplots in a 2x2 Grid')
plt.tight_layout()
plt.show()
````
